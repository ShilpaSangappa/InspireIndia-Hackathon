```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline  

pd.options.display.max_columns = None
ccms_data = pd.read_excel('/home/sindhuvarun/EELS hackathon/CCMS/Dummy data/Dummy data/Instant Raw Report (11).xlsx')
```

```python
timestamp = ccms_data['Timestamp']
energy = ccms_data['Total Apparent Energy']
print(timestamp)
```

```python
timestamp_split = timestamp.str.split(' ', expand=True)
date_frame = timestamp_split[0].str.split('-', expand=True)
time_frame = timestamp_split[1].str.split(':', expand=True)
#timestamp[0].str.split('-', expand=True)
#year, month, date = timestamp.str.split('-')[0], timestamp.str.split('-')[1], timestamp.str.split('-')[2]
```

```python
features = pd.concat([energy, date_frame, time_frame], axis=1)
```
